# Get groups

```julia
using JSON
neuronIdList = JSON.parsefile(joinpath(DATASET_DIR, "neuron_ids.json"))["analysis"] |> Vector{Int}
```

```julia
using HDF5

clusters = h5read("$(DATASET_DIR)/clusters.h5", "clusters") |> Vector{Int};
# make sure that the cluster Id start from 1 rather than 0
if minimum(clusters)==0
    clusters .+= 1
end
@assert minimum(clusters)==1
```

```julia
numGroups = unique(clusters) |> length


groups = Vector{Int}[]
for i in 1:numGroups
    push!(groups, Int[])
end

for (i, neuronId) in enumerate(neuronIdList)
    groupId = clusters[i]
    push!(groups[groupId], neuronId)
end

for (groupId, group) in enumerate(groups)
    println("$(groupId): $(group)\n")
end
```
