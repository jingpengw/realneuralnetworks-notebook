# Hierarchical clustering

```python
import h5py
import numpy as np
with h5py.File(os.path.join(DATASET_DIR, 'data.h5')) as f:
    neuron_id_list = np.asarray(f['neuronIdList'])
    similarity_matrix = np.asarray(f['nblast/zfishTable/meanSimilarityMatrix'])
    clusters = np.asarray(f['nblast/zfishTable/clusters'])
    orders = np.asarray(f['nblast/zfishTable/orders'])
    ordered_clusters = np.asarray(f['nblast/zfishTable/orderedClusters'])
    ordered_neuron_id_list = np.asarray(f['nblast/zfishTable/orderedNeuronIdList'])
    
```

```python
from scipy.spatial.distance import squareform

from scipy.cluster.hierarchy import dendrogram, linkage, cut_tree
import fastcluster as fc
from matplotlib import pyplot as plt

distance_matrix = - similarity_matrix - np.min(-similarity_matrix)
# transform redundent distance matrix to condensed matrix
y = squareform(distance_matrix)

linked = fc.ward(y)
ct = cut_tree(linked, n_clusters=26).flatten()

dend = dendrogram(linked)

plt.show()
```

```python
ordered_neuron_ids = neuron_id_list[dend['leaves']]
ordered_group_ids = ct[dend['leaves']]

group_id2neuron_ids = dict()

import pandas as pd
for group_id in pd.unique(ordered_group_ids):
    group_id2neuron_ids[group_id] = []
    
for i,group_id in enumerate(ordered_group_ids):
    neuron_id = ordered_neuron_ids[i]
    group_id2neuron_ids[group_id].append(neuron_id)
    
for group_id,neuron_ids in group_id2neuron_ids.items():
    print('group {}: {}\n'.format(group_id, neuron_ids))
    
```

```python
import scipy.cluster.hierarchy as hc
print('the two cluster is the isomorphic? ', hc.is_isomorphic(ct,clusters))

from sklearn.metrics import adjusted_rand_score

adjusted_rand_score(clusters, ct)
```
