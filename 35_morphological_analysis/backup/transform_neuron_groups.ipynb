```python
import json
import os

DATASET_DIR = "../01_data/"
```

```python
with open(os.path.join(DATASET_DIR, 'atlas/nblast/neuron_groups.json')) as f:
    neuron_groups = json.load(f)

N = len(neuron_groups)

d = dict()
for gid, group in enumerate(neuron_groups):
    if gid == N-1:
        gid = 'outliers'
        
    d[f'{gid}'] = group

```

```python
identified_neurons = [77931,77456,77260,76562,77266,77264,77261,77267,77268,77263,77265,77262,76202,77929,77259]


identified_neurons_set= set(identified_neurons)

for gid, group in d.items():
    group = set(group).difference(identified_neurons_set)
    
    d[gid] = list(group)
    
d['anatomically_identified'] = identified_neurons

d
        
```

```python
with open(os.path.join(DATASET_DIR, 'atlas/nblast/neuron_groups.json'), 'w') as f:
    json.dump(d, f, indent=2)
```
